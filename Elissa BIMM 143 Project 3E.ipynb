# Scientific Question: Do mutations in the ORF of the SARS-CoV-2 glycoprotein increase infectivity?

The virality of coronavirus SARS-CoV-2 virus causing COVID-19 disease is due to the presence of a surface glycoprotein, or "spike" protein that attaches to homosapien ACE2 receptors. Although SARS-CoV-2 is an mRNA virus, it mutates slowly than other RNA viruses. With that being said, some mutations have still occured, and mutations within the glycoprotein ORF can affect the infectivity of the virus. Specifically, the D614G isolate has a D --> G amino acid change at the 614th codon. This single AA change causes a decrease in S1 shedding (Zhang).  

# Hypothesis: A nucleotide mutation that leads to an amino acid change can increase infectivity, especially if the amino acid is a change from D to G.
